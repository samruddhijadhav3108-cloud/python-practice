# **IF-ELIF-ELSE STATEMENT**

# ***Q1. Positive, Negative or Zero***

```python
num = int(input("Enter a number: "))
if num > 0:
  print("Positive")
elif num < 0:
  print("Negative")
else:
  print("Zero")
```

# ***Q2. Grade Based on Marks***

```python
marks = int(input("Enter Marks: "))
if marks >= 90:
  print("Grade A")
elif marks >= 75:
  print("Grade B")
elif marks >=50:
  print("Grade C")
else:
  print("Fail")
```

# ***Q3. Traffic light signal***

```python
color = input("Enter traffic light color: ")
if color == "Red":
  print("Stop")
elif color == "Yellow":
  print("Wait")
elif color == "Green":
  print("Go")
else:
  print("Invalid color")
```

# ***Q4. Day name from number***

```python
day = int(input("Enter day number (1-7): "))
if day == 1:
  print("Monday")
elif day == 2:
  print("Tuesday")
elif day == 3:
  print("Wednesday")
elif day == 4:
  print("Thrusday")
elif day == 5:
  print("Friday")
elif day == 6:
  print("Saturday")
elif day == 7:
  print("Sunday")
else:
  print("Invalid day")
```

# ***Q5. Largest of three numbers***

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))

if a > b and a > c:
  print("Largest:", a)
elif b > c:
  print("Largest: ", b)
else:
  print("Largest: ", c)
```

# ***Q6. Age Category***

```python
age = int(input("Enter Age: "))
if age < 13:
  print("Child")
elif age < 20:
  print("Teenager")
elif age < 60:
  print("Adult")
else:
  print("Senior Citizen")
```

# ***Q7. Season based on month***

```python
month = int(input("Enter month number: "))

if month in [12, 1, 2]:
  print("Winter")
elif month in [3, 4, 5]:
  print("Summer")
elif month in [6, 7, 8]:
  print("Monsoon")
elif month in [9, 10, 11]:
  print("Autumn")
else:
  print("Invalid month")
```

# ***Q8. Eletricity bill slab***

```python
units = int(input("Enter units: "))

if units <= 100:
  print("Bill = ₹0")
elif units <= 200:
  print("Bill = ₹500")
elif units <= 300:
  print("Bill = ₹1000")
else:
  print("Bill = ₹1500")
```

# ***Q9. BMI category***

```python
bmi = float(input("Enter BMI: "))

if bmi < 18.5:
  print("Underweight")
elif bmi < 25:
  print("Normal")
elif bmi < 30:
  print("Overweight")
else:
  print("Extremly overweight")
```
