# **Practical 4: Sort List without Using Built-in Functions**

Write a Python program to sort a list of numbers in ascending order without using any built-in sorting functions.

```python
numbers = [5, 2, 8, 1, 4]

for i in range(len(numbers)):
    for j in range(i + 1, len(numbers)):
        if numbers[i] > numbers[j]:
            numbers[i], numbers[j] = numbers[j], numbers[i]

print("Sorted list:", numbers)
```

# **Practical 5: Count Frequency of Elements in a List**

Write a Python program to count the frequency of each element in a list and display the result.

```python
numbers = [1, 2, 2, 3, 3, 3]

frequency = {}

for i in numbers:
    if i in frequency:
        frequency[i] += 1
    else:
        frequency[i] = 1

print("Frequency of elements:", frequency)
```

# **Practical 10: Word Frequency Counter Using Dictionary**

Write a Python program to count the frequency of each word in a given sentence using a dictionary.




```python
sentence = "python is easy and python is fun"

words = sentence.split()

frequency = {}

for i in words:
    if i in frequency:
      frequency[i] += 1
    else:
      frequency[i] = 1

print("Word frequency:", frequency)
```

# **Practical 15: Function to Check Prime Number**

Write a Python program to create a function that checks whether a given number is prime or not.



```python
num = 7
count = 0

for i in range(1, num + 1):
    if num % i == 0:
        count += 1

if count == 2:
  print("Prime")
else:
  print("Not Prime")
```

# **Practical 16: Recursive Function for Factorial**

Write a Python program to find the factorial of a number using recursion.



```python
def fact(n):
  if n==1:
    return 1
  else:
    return n*fact(n-1)

print(fact(5))
```

# **Practical 17: Recursive Fibonacci Series**

Write a Python program to generate the Fibonacci series using recursion.



```python
def fib(n):
  if n<=1:
    return n
  else:
    return fib(n-1)+fib(n-2)

for i in range (6):
  print(fib(i), end=" ")
```

# **Practical 19: Map, Filter, Reduce Example**

Write a Python program demonstrating the use of map(), filter(), and reduce() functions on a list of numbers.



```python
from functools import reduce
numbers = [1, 2, 3, 4]

sq = list(map(lambda x: x*x, numbers))
print("Map:",sq)

even = list(filter(lambda x: x%2==0, numbers))
print("Filter:",even)

sum = reduce(lambda x, y: x+y, numbers)
print("Reduce:",sum)
```

# **Practical 22: Count Vowels and Consonants**
