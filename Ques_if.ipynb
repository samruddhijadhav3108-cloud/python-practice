# **IF STATEMENT**

# ***Q1. Check if a number is positive.***

```python
num = int(input("Enter a number: "))
if num > 0:
  print("Number is positive")
```

# ***Q2. Check if a number is greater than 100***

```python
num = int(input("Enter a number: "))
if num > 100:
  print("Number is greater than 100")
```

# ***Q3. Check if a number is even.***

```python
num = int(input("Enter a number: "))
if num % 2 == 0:
  print("Number is even")
```

# ***Q4. Check if a character is a vowel.***

```python
ch = input("Enter a character: ").lower()
if ch in "aeiou":
  print("It is a vowel")
```

# ***Q5. Check if a number is divisible by 5***

```python
num = int(input("Enter a number: "))
if num % 5 == 0:
  print("Divisible by 5")
```

# ***Q9. Check if a year is leap year.***

```python
year = int(input("Enter year: "))
if (year % 4 == 0 and year % 100 != 0) or (year % 400 ==0):
  print("Leap year")
```
