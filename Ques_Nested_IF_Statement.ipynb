# ***Q1. Positive and Even Number***

```python
num = int(input("Enter number: "))

if num > 0:
  if num % 2 == 0:
    print("Positive and Even")
  else:
    print("Positive but Odd")
else:
  print("Not Positive")
```

# ***Q2. Pass and Distinction***

```python
marks = int(input("Enter marks: "))

if marks >= 40:
  if marks >= 75:
    print("Pass with Distinction")
  else:
    print("Pass")
else:
  print("Fail")
```

# ***Q3. Leap year and century year***

```python
year = int(input("Enter year: "))
if year % 100 == 0:
  if year % 400 == 0:
    print("Century Leap Year")
  else:
    print("Century but not Leap Year")
else:
  if year % 4 == 0:
    print("Leap Year")
  else:
    print("Not a Leap Year")
```

# ***Q4. Divisible by 3 and 5***

```python
num = int(input("Enter number: "))

if num % 3 == 0:
  if num % 5 == 0:
    print("Divisible by both 3 and 5")
  else:
    print("Divisible by 3 only")
else:
  print("Not divisible by 3")
```

# ***Q5. Username and password check***

```python
username = input("Enter usename: ")
password = input("Enter password: ")

if username == "admin":
  if password == "1234":
    print("Login Successful")
  else:
    print("Wrong Password")
else:
  print("Wrong Username")
```

# ***Q6. Valid and right-angled triangle***

```python
a = int(input("Enter side 1: "))
b = int(input("Enter side 2: "))
c = int(input("Enter side 3: "))

if ((a*a + b*b) == c*c):
  print("It is Right-angled triangle ")
else:
   print("It is  not Right-angled triangle ")
```

# ***Q7. Eligible voter and registered***

```python
age = int(input("Enter age: "))
registered = input("Registered (yes/no): ")

if age >=18:
  if registered == "yes":
    print("Eligible to vote")
  else:
    print("Not registered")
else:
  print("Under age")
```

# ***Q8. Three-digit palindrome***

```python
num = int(input("Enter Number: "))
if 100 <= num <= 999:
  if str(num) == str(num)[::-1]:
    print("Three-digit Palindrome")
  else:
    print("Not Palindrome")
else:
  print("Not three-digit number")
```

# ***Q9. Alphabet and uppercase***

```python
ch = input("Enter character: ")

if ch.isalpha():
  if ch.isupper():
    print("Uppercase Alphabet")
  else:
    print("Lowercase Alphabet")
else:
  print("Not an alphabet")
```

# ***Q10. Bank withdrawl through atm check***

```python
balance = 5001
amount = int(input("Enter withdrawl amount: "))

if amount <= balance:
  if amount % 100 == 0:
    print("Withdrawl Successful")
  else:
    print("Amount must be multiple of 100")
else:
  print("Insufficient Balance")
```
