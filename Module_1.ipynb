# **15. Use isinstance() function**

```python
# Using isinstance() in Python (with output)

x = 10
y = 5.5
name = "Samruddhi"

print(isinstance(x, int))
print(isinstance(y, float))
print(isinstance(name, str))
print(isinstance(x, (int, float)))
print(isinstance(name, int))
```

# **19. Identify Mutable vs Immutable types**





```python
# Mutable example (list)
numbers = [1, 2, 3]
numbers[0] = 10
print(numbers)

# Immutable example (string)
name = "Sam"
name = name + "ruddhi"
print(name)
```
