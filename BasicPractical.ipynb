# **17. Reverse a Number**

```python
n = 1234
rev = 0
temp = n # Keep a temporary copy of n for the loop
while temp > 0:
  digit = temp % 10
  rev = rev * 10 + digit
  temp = temp // 10
print("Original number:", n)
print("Reversed number:", rev)
```

# **18. Check Palindrome Number**

```python
n = 12345
original_n = n  # Store the original number
rev = 0

while n > 0:
  digit = n % 10
  rev = rev * 10 + digit
  n = n // 10

if original_n == rev:
  print("palindrome")
else:
  print("not palindrome")
```

# **19. Count digits in a number**

# **25. Simple Calculator using if-else**

```python
a = 10
b = 5
op = "+"

if op == "+":
  print(a + b)
elif op == "-":
  print(a - b)
elif op == "*":
  print(a * b)
elif op == "/":
  print(a / b)
else:
  print("Invalid operator")
```
