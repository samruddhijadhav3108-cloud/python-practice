```python
def env_info():
  R = input("Name the Room A or B: ")
  S = input("Name the Status Clean or Dirty: ")
  return (R,S)

```

```python
def agent_action(Room, Status):
  if (Room == "A" and Status == "Clean"):
    return "Go to room B"
  elif (Room == "B" and Status == "Clean"):
    return "Go to room A"
  elif (Room == "A" and Status == "Dirty"):
    return "Suck Dirt"
  elif (Room == "B" and Status == "Dirty"):
    return "Suck Dirt"
```

```python
agent_action(*env_info())
```
