# **1. Write a function to find factorial of a number.**

```python
def factorial(n):
    fact = 1
    for i in range(1, n+1):
        fact *= i
    return fact

print(factorial(5))
```

# **2. Function to check palindrome number**

```python
def palindrome(num):
    if str(num) == str(num)[::-1]:
        return "Palindrome"
    else:
        return "Not Palindrome"

print(palindrome(121))
```

# **3. Function to find sum of digits**

```python
def sum_digits(num):
    total = 0
    for i in str(num):
        total += int(i)
    return total

print(sum_digits(123))
```

# **21. Prime Number**

```python
num = 7
is_prime = True

for i in range(2, num):
    if num % i == 0:
        is_prime = False
        break

if is_prime and num > 1:
    print("Prime")
else:
    print("Not Prime")
```

# **22. Armstrong Number (153 example)**

```python
num = 153
temp = num
total = 0

while temp > 0:
    digit = temp % 10
    total += digit ** 3
    temp //= 10

if total == num:
    print("Armstrong")
else:
    print("Not Armstrong")
```

# **32. Count Vowels**

```python
def count_vowels(text):
    count = 0
    for ch in text:
        if ch.lower() in "aeiou":
            count += 1
    return count

print(count_vowels("Samruddhi"))
```

# **34. Largest in List (without max)**

```python
def find_largest(lst):
    largest = lst[0]
    for num in lst:
        if num > largest:
            largest = num
    return largest

print(find_largest([4, 9, 2, 15, 6]))
```

# **37. Circle Area**

```python
class Circle:
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius

c = Circle(5)
print(c.area())
```

# **39. BankAccount**

```python
class BankAccount:
    def __init__(self, balance):
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        self.balance -= amount

acc = BankAccount(1000)
acc.deposit(500)
acc.withdraw(200)
print(acc.balance)
```

# **40. Student Average**

```python
class Student:
    def __init__(self, marks):
        self.marks = marks

    def average(self):
        return sum(self.marks) / len(self.marks)

s = Student([80, 90, 70])
print(s.average())
```
