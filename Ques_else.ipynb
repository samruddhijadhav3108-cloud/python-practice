# **IF-ELSE STATEMENT**

# ***Q1. Positive or negative number.***

```python
num = int(input("Enter a number: "))
if num >= 0:
  print("Positive number")
else:
  print("Negative number")
```

# ***Q2. Even or odd number.***

```python
num = int(input("Enter a number: "))
if num % 2 == 0:
  print("Even number")
else:
  print("Odd number")
```

# ***Q3. Vote or not***

```python
age = int(input("Enter age: "))
if age >= 18:
  print("Can vote")
else:
  print("Cannot vote")
```

# ***Q4. Greater than 50 or not***

```python
num = int(input("Enter a number: "))
if num > 50:
  print("Greater than 50")
else:
  print("Not greater than 50")
```

# ***Q5. Pass or fail***

```python
marks = int(input("Enter a number: "))
if marks >= 35:
  print("Pass")
else:
  print("Fail")
```

# ***Q6. Divisible by 3 or not***

```python
num = int(input("Enter a number: "))
if num % 3 == 0:
  print("Divisible by 3")
else:
  print("Not divisible by 3")
```

# ***Q7. Vowel or consonant***

```python
ch = input("Enter a character: ").lower()
if ch in "aeiou":
  print("Vowel")
else:
  print("Consonant")
```

# ***Q8. Leap year or not***

```python
year = int(input("Enter year: "))
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
  print("Leap year")
else:
  print("Not a leap year")
```

# ***Q9. Password valid or not***

```python
password = input("Enter password: ")
if password == "admin123":
  print("Password is valid")
else:
  print("Password is invalid")
```

# ***Q10. Zero or non-zero***

```python
num = int(input("Enter a number: "))
if num == 0:
  print("Zero")
else:
  print("Non-zero")
```
