# Distance Formula
d = √((x₁ – x₂)² + (y₁ – y₂)²)

```python
import math

x1 = float(input("Enter x1: "))
y1 = float(input("Enter y1: "))
x2 = float(input("Enter x2: "))
y2 = float(input("Enter y2: "))

distance = math.sqrt((x1 - x2)**2 + (y1 - y2)**2)

print("Distance =", distance)
```

# Slope Formula
Slope = (y₂ – y₁) / (x₂ – x₁)

```python
x1 = float(input("Enter x1: "))
y1 = float(input("Enter y1: "))
x2 = float(input("Enter x2: "))
y2 = float(input("Enter y2: "))

slope = (y2 - y1) / (x2 - x1)

print("Slope =", slope)
```

# CODE

```python
# Area of Rectangle
def area_rectangle(l, b):
    return l * b

# Perimeter of Rectangle
def perimeter_rectangle(l, b):
    return 2 * (l + b)

# Area of Square
def area_square(a):
    return a * a

# Perimeter of Square
def perimeter_square(a):
    return 4 * a

# Area of Circle
def area_circle(r):
    return 3.14 * r * r

# Circumference of Circle
def circumference_circle(r):
    return 2 * 3.14 * r

#Area of Triangle
def area_triangle(b, h):
    return 1/2 * b * h

#Pythagorean Theorem
def pythagorean(a, b):
    return (a*a + b*b)

# Average
def average(values):
    return sum(values) / len(values)

# Quadratic Formula
def quadratic(a, b, c):
    discriminant = b**2 - 4*a*c
    if discriminant >= 0:
        root1 = (-b + math.sqrt(discriminant)) / (2*a)
        root2 = (-b - math.sqrt(discriminant)) / (2*a)
        return root1, root2
    else:
        # Complex roots
        root1 = (-b + math.sqrt(discriminant)) / (2*a)
        root2 = (-b - math.sqrt(discriminant)) / (2*a)
        return root1, root2

# Probability
def probability(favorable, total):
    return favorable / total

# Distance Formula
def distance(x1, y1, x2, y2):
    return math.sqrt((x1 - x2)**2 + (y1 - y2)**2)

# Slope Formula
def slope(x1, y1, x2, y2):
    if x2 == x1:
        return "undefined"
    return (y2 - y1) / (x2 - x1)

# Slope-Intercept Formula
def slope_intercept(m, b, x):
    return m * x + b

# Complex Number Standard Form
def complex_standard(a, b):
    return complex(a, b)

# Modulus of Complex Number
def complex_modulus(a, b):
    return math.sqrt(a**2 + b**2)

# Conjugate of Complex Number
def complex_conjugate(a, b):
    return complex(a, -b)

# Volume of Cube
def volume_cube(a):
    return a**3

# Volume of Cylinder
def volume_cylinder(r, h):
    return math.pi * r**2 * h

# Surface Area of Sphere
def surface_area_sphere(r):
    return 4 * math.pi * r**2

# Sine (SOH)
def sine(opposite, hypotenuse):
    return opposite / hypotenuse

# Cosine (CAH)
def cosine(adjacent, hypotenuse):
    return adjacent / hypotenuse

# Tangent (TOA)
def tangent(opposite, adjacent):
    return opposite / adjacent

print("Area of Rectangle:", area_rectangle(10, 5))
print("Perimeter of Rectangle:", perimeter_rectangle(10, 5))
print("Area of Square:", area_square(4))
print("Perimeter of Square:", perimeter_square(4))
print("Area of Circle:", area_circle(7))
print("Circumference of Circle:", circumference_circle(8))
print("Area of Triangle:", area_triangle(10, 5))
print("Pythagorean Theorem (3,4):", pythagorean(3, 4))
print("Average:", average([10, 20, 30]))
print("Quadratic roots (1,-3,2):", quadratic(1, -3, 2))
print("Probability (favorable 3, total 10):", probability(3, 10))
print("Distance (2,3) to (5,7):", distance(2,3,5,7))
print("Slope (2,4) to (6,8):", slope(2,4,6,8))
print("Slope-Intercept y=2x+1 for x=3:", slope_intercept(2, 1, 3))
print("Complex number standard form (3,4):", complex_standard(3,4))
print("Complex modulus (3,4):", complex_modulus(3,4))
print("Complex conjugate (3,4):", complex_conjugate(3,4))
print("Volume of Cube (4):", volume_cube(4))
print("Volume of Cylinder (r=3,h=5):", volume_cylinder(3,5))
print("Surface Area of Sphere (r=5):", surface_area_sphere(5))
print("Sine (opposite 3, hypotenuse 5):", sine(3,5))
print("Cosine (adjacent 4, hypotenuse 5):", cosine(4,5))
print("Tangent (opposite 3, adjacent 4):", tangent (3,4))
```
